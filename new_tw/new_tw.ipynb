Мы исследовали решение на сепаратрисе в круглой трубе. Если длина расчетной области достаточно большая, то решение оказывается локализованным в пространстве длиной около 20R. В статье [Chantry, Willis, Kerswell, Genesis of Streamwise-Localized Solutions from Globally Periodic Traveling Waves in Pipe Flow, PRL 112, 164501 (2014)] показано, как наше локализованное решение в трубе возникает в результате бифуркации потери симметрии из бегущей волны. Чтобы это продемонстрировать, они, пользуясь методом Ньютона, продлили локализованное решение в сторону уменьшая длины расчетной области. В результате пришли к трем периодам бегущей волны. Вероятно, это и есть та бегущая волна, которая возникает на сепаратрисе в короткой расчетной области. Чтобы это гарантировать, нужно повторить их результат. Я пробовал это сделать, но не довел дело до конца. Тем не менее, каких-то принципиальных сложностей нет, и можно довести. Этот факт еще раз показывается связь между бегущей волной и локализованным решением, которое мы исследовали. 

В этой же статье было показало, что решение на сепаратрисе оказывается периодическим не только с периодичностью по углу pi, но и с периодичностью по углу 2pi/3, но при немного больших значениях числа Рейнольдса, 2400, по-моему. 

Решение на сепаратрисе обладает существенным недостатком. В фазовом пространстве оно принадлежит границе области притяжения турбулентного режима течения, находится в стороне от турбулентного аттрактора. (Хотя использовать слово “аттрактор” не совсем корректно, я буду делать это здесь для краткости. Возможно было бы лучше “седло”.) Реальная турбулентность никогда не будет близка к решению на сепаратрисе. Это не значит, что выводы, полученные для решения на сепаратрисе, неприменимы для турбулентности, но они были бы убедительнее, если бы были проверены и на других простых решениях. Решение на сепаратрисе можно непрерывным образом продлить в пространстве параметров, и перейти на верхнюю ветвь решений. Решения с верхней ветви имеют большую амплитуду пульсаций, меньшую скорость сноса, и находятся уже внутри области притяжения турбулентности. (Это стоило бы продемонстрировать, например, показав, что любые возмущения такого решения ведут к развитию турбулентной.) Тем не менее это не гарантирует того, что реальная турбулентность приближается к такому решению. 

Во время исследования механизма самоподдержания в круглой трубе возникло желания получить также бегущую волну в плоском канале. Это позволит получить более убедительные результаты: исключить влияние кривизны стенки,  ввести ортогональную систему координат, упростив тем самым выкладки. Оказывается, что решение на сепаратрисе в плоском канале имеет новую форму. Оно  прижато к противоположной относительно твердой стенки границе, и если увеличивать расчетную область, то можно ожидать, что оно будет удаляться от стенки на бесконечность. По видимому в круглой трубе решение на сепаратрисе похоже на пристенное решение только по тому, что расчетная область сужается по мере удаления от стенки, и решение на сепаратрисе не может отдалиться от нее на значительное расстояние. В тоже время, уже в круглой трубе было замечено, что с увеличением числа Рейнольдса, решение на сепаратрисе приближается к оси трубы, а не к стенке, что можно интерпретировать так, что оно не является пристенным. ## Круглая труба

Сперва новая бегущая волна была найдена в круглой трубе при тех же граничных условиях и условиях симметрии, что и решение на сепаратрисе. 

### Постановка задачи

Рассмотрим течение жидкости в прямой трубе круглого сечения постоянного радиуса. Введем цилиндрическую систему координат $(x,r,\theta)$ - продольная, радиальная и угловая координаты. Считаем жидкость вязкой и несжимаемой, удовлетворяющей системе уравнения Навье-Стокса

$$
div \mathbf{v} = 0
$$

$$
\frac{\partial \mathbf v}{\partial t} = - (\mathbf v, \nabla) \mathbf v - \nabla p + \frac{1}{Re} \Delta \mathbf v
$$

В качестве масштаба длины выступает радиус трубы R, в качестве масштаба скорость - скорость на оси трубы в течении Пуазейля U. Если течение турбулентное, что масштаб скорости вводится как скорость на оси трубы в течении Пуазейля с тем же расходом. Таким образом число Рейнольдса определяется по формуле $Re = UR / \nu$. Значение числа Рейнольдса, введенного таким образом, совпадает со значением числа Рейнольдса, введенного через диаметр D = 2R и расходную скорость Uq = U/2. 

Жидкость приводится в движение продольным градиентом давления, определяемым из условия постоянства средней скорости (определенной по расходу) Uq = 1/2 в безразмерных единицах измерения. 

В безразмерных единицах измерения радиус трубы равен 1. На стенках трубы ставится условие прилипания $\mathbf v = 0$ при $r=1$. На оси трубы скорость ограничена. Будем искать решения, периодические вдоль трубы и в угловом направлении с периодами L и $2 \pi / n$ соответственно

$$
X : \mathbf v (x,r,\theta) = \mathbf (x+L, r, \theta)
$$

$$
\Omega _n : \mathbf v (x,r,\theta) = \mathbf v (x,r,\theta + 2\pi/n)
$$

Если L может быть любым действительным положительным число, то из физических соображений n может принимать только натуральные значения, однако в расчетах бывает удобно рассматривать действительные значения n тоже. 

На поле скорости $\mathbf v = (u,v,w)$ накладывается дополнительное условие зеркальной симметрии относительно плоскости диаметрального сечения $\theta = 0$

$$
Z: (u,v,w)(x,r,\theta) = (u,v,-w) (x,r,-\theta)
$$

Это условие было поставлено в работе [Avila, M., Mellibovski, F., Roland, N. & Hof, B. 2013 Streamwise-localized solutions at the onset of turbulence in pipe flow. Phys. Rev. Lett. 110, 224502.] и необходимо для того, чтобы решение на сепаратрисе в протяженной расчетной области оказалось не только локализованным в пространстве, но и периодическим по времени. При нахождении новой бегущей волны оно также играет ключевую роль, как будет показано ниже. 

### Численный метод

Работа выполняется численно, методом прямого численного интегрирования. Для этого уравнения Навье-Стокса для несжимаемой жидкости решаются конечно-разностным методом [], объединяющим пространственную аппроксимацию второго порядка точности и полу-неявный метод Рунге-Кутта третьего порядка точности для интегрироваться по времени. Основные результаты были получены в расчетной области длиной L=5 при n=2, таким образом, в силу условий симметрии $\Omega_2$ и $Z$, длина расчетной области в угловом направлении составляет $\pi/2$. Расчетная сетка включает 64 точки в продольном направлении, 40 точек в радиальном направлении, со сгущением у твердой стенке, 32 точки в угловом направлении на четверть трубы. 

### Метод Ньютона-Крылова

Поле скорости бегущей волны удовлетворяет следующему соотношению

$$
\mathbf v (x,r,\theta,t) = \mathbf v_0 (x-ct,r,\theta)
$$

где $c$ - фазовая скорость волны. 

В случае, если движение жидкости описывается уравнениями Навье-Стокса, поле скорости бегущей волны $\mathbf v$ удовлетворяет стационарному нелинейному уравнению

$$
- c \frac{\partial \mathbf v}{\partial x} = - (\mathbf v, \nabla) \mathbf v - \nabla p + \frac{1}{Re} \Delta \mathbf v
$$

Это уравнение может быть решено методом Ньютона. Метод Ньютона итерационный и требует начального приближения. Основная сложность при находении простых решений состоит в том, чтобы найти начальное приближение, достаточно близкое к некоторому решению, с которого метод сойдется к нему. 

Пусть в конечномерном представлении поле скорости определяется m независимыми переменными. Параметрами системы могут быть скорость сноса c, число Рейнольдса Re, длина расчетной области L и периодичность в угловом направлении n. 

Реализация метода Ньютона-Крылова описана в статье [Viswanath, D. 2007 Recurrent motions within plane couette turbulence. J. Fluid Mech. 580, 339–358]. Я не видел, чтобы кто-то ссылался на другие статьи. Метод, описанный в статье, немного сложнее, чем в моей реализации. 

С формальной точки зрения поле скорости v должно задаваться независимыми переменными. Ествественно представление поля скорости v, задаваемое зачением его компонент u,v,w в каждой точке стеки не удавлтворяет этому требованию, так как на них наложено дополнительно ограничение солиноидальности поля скорости. Висванаши предлагает перейти от етстветственно представления поля скорости к полям нормальной к стенке скорости и нормальной к стенке завихренности. Такой переход требует кардинальной перестройки кода программы. Тем не менее, в этом нет необходимости, и можно реализовать метод НьютонаКрылова так

Формально система недоопределена, и необходимо вводить дополнительные ограничения, чтобы система имела единственное решение. Это могут быть ограничения сдвига по времени и в пространстве. Построение подпространств Крылова с этими ограничения несколько неестветсвенно. По факту в этом нет необходимости, методы Крылова позволяют решать и недоопредененные системы (это стоит обосновать). Не вводя явным образом дополнительные ограничения сдвига в код, можно сохранить его гибкость в плане выбора определяемых параметров. 

### Получение бегущей волны

В локализованном решениии на сепаратрисе можно выделить бегущую вниз по потоку волну. При этом длина расчетной области на порядок больше, чем длина этой бегущей волны, равной примерно 5R. Это значит, что ее длина не навязана граничными условиями, а формируется естветвенным образом. Можно ожидать, что в реальном турбуленином движении бегущие волны будут иметь именно такую длину. 

При построении решения на сепаратрисе можно отметить следующею особенность. Те приближения к сепаратрисе, которые развиваются в турбулениное течение, уже после того, как покинут сепаратрису, но прежде, чем приобретут хаотический харрактер движения, выходят на некоторое время на режим с постоянной амплитудой возмущений. Возможно это верхняя ветвь решения, принадлежащего сепаратрисе, либо какое-то другое простое решения. Можно внимательнее изучить решение в этой области. Возможно удастся выбрать начальные данные для метода Ньютона, с которыми он сойдется, и даст таким образом новое простое решение. Для этого в расчтеной области длиной 5R, была проведена серия расчетов, в которых в качетве начальных данных было взято предварительно найденное решение на сепаратрисе, возмущенное на небольшую величину. В тех случаях, когда развивался турбунеиный режим течения, в нем выбирались моменты, близкие к бегущей волне. 

Можно не вдаваться в такие подробности при описании мотивации этого поступка, и сказать проще. Можно обратиться непосредственно к турбунеинотному течению жидкости, и проанализировать его на наличие бегущих волн.
