## Круглая труба

Сперва новая бегущая волна была найдена в круглой трубе при тех же граничных условиях и условиях симметрии, что и решение на сепаратрисе. 

### Постановка задачи

Рассмотрим течение жидкости в прямой трубе круглого сечения постоянного радиуса. Введем цилиндрическую систему координат $(x,r,\theta)$ - продольная, радиальная и угловая координаты. Считаем жидкость вязкой и несжимаемой, удовлетворяющей системе уравнения Навье-Стокса

$$
div \mathbf{v} = 0
$$

$$
\frac{\partial \mathbf v}{\partial t} = - (\mathbf v, \nabla) \mathbf v - \nabla p + \frac{1}{Re} \Delta \mathbf v
$$

В качестве масштаба длины выступает радиус трубы R, в качестве масштаба скорость - скорость на оси трубы в течении Пуазейля U. Если течение турбулентное, что масштаб скорости вводится как скорость на оси трубы в течении Пуазейля с тем же расходом. Таким образом число Рейнольдса определяется по формуле $Re = UR / \nu$. Значение числа Рейнольдса, введенного таким образом, совпадает со значением числа Рейнольдса, введенного через диаметр D = 2R и расходную скорость Uq = U/2. 

Жидкость приводится в движение продольным градиентом давления, определяемым из условия постоянства средней скорости (определенной по расходу) Uq = 1/2 в безразмерных единицах измерения. 

В безразмерных единицах измерения радиус трубы равен 1. На стенках трубы ставится условие прилипания $\mathbf v = 0$ при $r=1$. На оси трубы скорость ограничена. Будем искать решения, периодические вдоль трубы и в угловом направлении с периодами L и $2 \pi / n$ соответственно

$$
X : \mathbf v (x,r,\theta) = \mathbf (x+L, r, \theta)
$$

$$
\Omega _n : \mathbf v (x,r,\theta) = \mathbf v (x,r,\theta + \frac{2\pi}{n})
$$

Если L может быть любым действительным положительным число, то из физических соображений n может принимать только натуральные значения, однако в расчетах бывает удобно рассматривать действительные значения n тоже. 

На поле скорости $\mathbf v = (u,v,w)$ накладывается дополнительное условие зеркальной симметрии относительно плоскости диаметрального сечения $\theta = 0$

$$
Z: (u,v,w)(x,r,\theta) = (u,v,-w) (x,r,-\theta)
$$

Это условие было поставлено в работе [Avila, M., Mellibovski, F., Roland, N. & Hof, B. 2013 Streamwise-localized solutions at the onset of turbulence in pipe flow. Phys. Rev. Lett. 110, 224502.] и необходимо для того, чтобы решение на сепаратрисе в протяженной расчетной области оказалось не только локализованным в пространстве, но и периодическим по времени. При нахождении новой бегущей волны оно также играет ключевую роль, как будет показано ниже. 

Комбинация условия $\Omega_n$ и $Z$ дают условие зеркальной симметрии относительно диаметрального сечения $\theta = m\pi/n$, где m - целое чило. 

$$
Z_n: (u,v,w)(x,r,\frac{m\pi}{n} + \theta) = (u,v,-w)(x,r,\frac{m\pi}{n} - \theta)
$$

Таким образом можно считать, что рассчетная область заключена между плоскостями $\theta=0$ и $\pi/n$, а на стенках ставится условие зеркальной симметрии, эквиволентное условию равентва нулю нормальной скорости w, и производных по нормали от касательных компоненты скорости u,v, и давления p.

### Численный метод

Работа выполняется численно, методом прямого численного интегрирования. Для этого уравнения Навье-Стокса для несжимаемой жидкости решаются конечно-разностным методом [], объединяющим пространственную аппроксимацию второго порядка точности и полу-неявный метод Рунге-Кутта третьего порядка точности для интегрироваться по времени. Основные результаты были получены в расчетной области длиной L=5 при n=2, таким образом, в силу условий симметрии $\Omega_2$ и $Z$, длина расчетной области в угловом направлении составляет $\pi/2$. Расчетная сетка включает 64 точки в продольном направлении, 40 точек в радиальном направлении, со сгущением у твердой стенке, 32 точки в угловом направлении на четверть трубы. Вдоль трубы и в угловом направлении точки расположены на равном расстоянии друг от друга. 

### Метод получения бегущей волны

В данной постановке решение, сдвинутое вдоль трубы, остается решением. В этом случае возможно существование решений типа бегущей волны, удавлетворяющих уравнению

$$
\mathbf v (x,r,\theta,t) = \mathbf v_0 (x-ct,r,\theta)
$$

где $c$ - фазовая скорость волны. 

В случае, если движение жидкости описывается уравнениями Навье-Стокса, поле скорости бегущей волны $\mathbf v$ удовлетворяет стационарному нелинейному уравнению

$$
- c \frac{\partial \mathbf v}{\partial x} = - (\mathbf v, \nabla) \mathbf v - \nabla p + \frac{1}{Re} \Delta \mathbf v
$$

Это уравнение может быть решено методом Ньютона. Метод Ньютона итерационный и требует начального приближения. Основная сложность при находении простых решений состоит в том, чтобы найти начальное приближение, достаточно близкое к некоторому решению, с которого метод сойдется к нему.

### Общие соображения о бегущих вольных

Метод Ньютона позволяет находить бегущие волны и другие решения, удовлетворяющие заданным ограничениям, но он не сходится с произвольным начальным приближением. Начальное приближение должно быть достаточно близко к некоторому точному решению, тогда метод Ньютона позволит найти это решение. 

Все найденные к настоящему времени бегущие волны и периодические в подвижной системе отсчета решения являются неустойчивыми, и не могут быть наблюдаемы в эксперименте в течении длительного время. Тем не менее, число неустойчивых направлений может быть не велико, и такие решения играют важную роль в процессе поддержания турбулентного течения. В фазовом пространстве каждому состоянию течения соответствует точка, которая со временем описывает некоторую кривую. Бегущим волнам и периодическим решениям, при условии периодичности вдоль трубы, соответствуют некоторые замкнутые траектории в фазовом пространстве. Каждое решение притягивает к себе траекторию вдоль устойчивых направлений, и отталкивает вдоль неустойчивых. Уходя от одного решения вдоль неустойчивого направления, траектория может попасть в область притяжения другого решения, и приблизиться к нему. Таким образом бегущие волны и периодические по времени решения можно рассматривать, как скелет турбулентности. Можно ожидать, что чем выше устойчивость решения, тем больше времени турбулентность будет проводить вблизи него. 

Я не очень хорошо знаком с терминологией, и не очень точно изложил эту теорию, но суть надеюсь уловил. У этой теории есть некоторые проблемы, связанные например с тем, что решения, приближающие турбулентный аттрактор, меняются с изменением размеров расчетной области. В общем, с увеличением расчетной области необходимы все более сложные периодические решения со все большим периодом по времени. Тем не менее, решения, найденные в небольших расчетных областях, могут позволить получить турбулентную статистику, либо оказаться удобными моделями для исследования реальных турбулентных процессов, например процесса самоподдержания турбулентности. 

Наиболее естественный способ получить начальные приближения для метода Ньютона является выделение их из реальной турбулентности. Наблюдая за турбулентным течением в небольшой расчетной области можно попытаться оценить, какое время турбулентность проводит около какой-либо из бегущих волн, и выделить начальные данные для метода Ньютона. Мы считаем, что течение близко к бегущей волне, если форма его поля скорости практические не меняется во времени, но оно смещается в пространстве. Это исключительно качественное соображение. Чем больше времени турбулентность проводит около некоторого решения, тем выше шанс выделить его этим методом, таким образом этот метод позволяет получить наиболее важные простые решения. Существуют некоторые методы, позволяющие оптимизировать этот процесс. 

### Получение "новой" бегущей волны

Чтобы оценить возможность получения начального приближения к бегущей волне для метода Ньютона из турбулентного течения, мы воспроизвели турбулентное течения в расчетной области небольшого размера. 

Можно отметить еще одну деталь. При построении решения на сепаратрисе те приближения к сепаратрисе, которые развиваются в турбулентное течение, уже после того, как покинут сепаратрису, но прежде, чем приобретут хаотический характер движения, выходят на некоторое время на режим с постоянной амплитудой возмущений. Это может быть одно из простых решений, к которому приближается турбулентное движение, например верхняя ветвь решения, принадлежащего сепаратрисе. Если решение действительно приближается к некоторому простому решению, то возможно его удастся выделить методом Ньютона. По этой причине в качестве начальных значений для расчета взято решение на сепаратрисе, возмущенное на небольшую величину, развивающееся в турбулентность. 

Расчет проводился при условии $\pi$-периодичности по углу и периодичности вдоль трубы с периодом L=5. Учитывая условие зеркального отражения относительно диаметрального сечения Z, размер расчетной области в угловом направлении составил $\pi/2$. Длина расчетной области выбрана из следующих соображений. 

В локализованном решение на сепаратрисе можно выделить бегущую вниз по потоку волну. При этом длина расчетной области на порядок больше, чем длина этой бегущей волны, равной примерно 5R. Это значит, что ее длина не навязана граничными условиями, а формируется естественным образом. Можно ожидать, что в реальном турбулентном движении бегущие волны будут иметь именно такую длину.  

Оказалось, что при числе Рейнольдса Re=2200, турбулентное течение оказывается близко к бегущей волне. Оно существует во времени неограниченно долго. Во всяком случае я не наблюдал случаев, чтобы решение ламинаризовалось или покинуло это бегущую волну. Турбулентное поле скорости при этом оказалось достаточно хорошим приближением к решению на сепаратрисе, с которым метод Ньютона сходится и позволяет получить бегущую волну, вокруг которой организовано это движение. 

Ключевой причиной, по которой турбулентность оказывается близка к бегущей волне, по видимому, являются условие отражения относительно Z.  Оно каким-то образом поддерживает устойчивость простых решений. В достаточно протяженной расчетной области именно это условие позволяет получить периодическое локализованное решение на сепаратрисе. 

Бегущая волна, возникающая на сепаратрисе, которой был инициализирован расчет, обладает зеркальной симметрией относительно сечения $\theta = \pi/2n$ со сдвигом на половину периода вдоль трубы 

$$
 (u,v,w)(x,r,\frac{\pi}{2n} + \theta) = (u,v, - w)(x+\frac{L}{2},r,\frac{\pi}{2n} - \theta)
$$

При этом в расчетах эта симметрия сохраняется на протяжении значительного промежутка времени. Возможно, если начинать расчет с других начальных условий, не обладающих указанной симметрией, то турбулентный режим окажется другим. Возможно, но маловероятно. Найденный турбулентный режим является аттрактором, турбулентное движение, однажды попав в область притяжения найденной бегущей волны, оставаться вблизи нее навсегда. Если существуют какие-то другие режимы течения, то даже небольшой шанс того, что траектория может покинуть этот режим, ведет к тому, что с течением времени траектории его покинут и притянутся к аттрактору. 

Вероятно, для любых начальных данных турбулентный режим устанавливается только один, организованный вокруг найденной бегущей волны. Это стоит проверить. 

Очевидно, что существует некоторый диапазон параметров, в котором турбулентность оказывается близка к бегущей волне. Можно попытаться его определить. 

Картинки и описание можно найти в файле https://nbviewer.jupyter.org/github/pimanov/research/blob/master/production/nk/tw2200new/newTW-Re2200.ipynb, а также описание метода Ньютона-Крылова.

## Плоский канал

Метод, описанный выше, обладает некоторой универсальностью, и позволяет получить бегущую волну не только в трубе, но и в плоском канале. Для этого достаточно ввести аналогичные ограничения симметрии в плоском канале. 

### Постановка задачи

Считаем, что течение жидкости несжимаемым, удовлетворяющим системе уравнений Навье-Стокса.

Рассмотрим течение жидеости между двумя параллельными плоскостями. Жидкость приводится в движение внешним градиентом давления, стенки покоятся. Введем условие симметрии S поля скорости относительно средней плоскости, параллельной твердым стенкам. Это условие эквивалентно условию проскальзывания не средней плоскости. Это стандартное условие, применемое при расчете течений в плоских каналах, позволяющее исключить влияние одной из стенок на процессы, протикающие на противоположной стенке. В данном случае оно эквивалетно условию на оси в круглой трубе, вытекающее из условия зеркальной симметрии относительно диаметрального сечения. 

В качестве характерных единиц измерения выберем расстояние от стенки до центральной плоскости H, максимальную скорость в течении Пуазейля U, и кинематическую вязкость $\nu$. Если характер течения отличается от ламинарного, то характерная скорость определяется, как максимальная скорость в течении Пуазейля, имеющем тот же расход. Число Рейнольдса $Re = UH / \nu$. Число Рейнольдса, введенное таким образом, отличается от числа Рейнольдса $Re_q = U_q D / \nu$, введенного через расходную скорость $U_q = 2U/3$ и расстояние меду стенками $D = 2H$. $Re_q = 4/3Re$

Введем афинную систему координат (x,y,z). Ось x направлена вдоль внешнего потока, параллельна стенке. Ось y направлена по нормали к стенке, на стенке y=0, в центре расчетной области y=1. Ось z направлена поперек потока, параллельно стенкам. 

Условие отражения относительно центральной плоскости выражается уравнением

$$
S: (u,v,w)(x,1-y,z) = (u,-v,w)(x,1+y,z)
$$

По аналогии с постановкой задачи в круглой трубе введем условие периодичности в продольном и трансверсальном направлениях с периодами $L_x$ и $L_z$, а также условие зеркальной симметрии относительно плоскости $z = 0$. 

$$
T_x: \mathbf v (x,y,z) = \mathbf v (x+L_x,y,z)
$$

$$
T_z: \mathbf v (x,y,z) = \mathbf v (x,y,z+L_z)
$$

$$
Z: (u,v,w) (x,y,z) = (u,v,-w) (x,y,-z)
$$

Условия $T_z$ и $Z$ дают условие зеркальной симметрии относительно плоскости $z=L_z/2$

$$
Z_z: (u,v,w) (x,y,L_z/2 + z) = (u,v,-w) (x,y,L_z/2 - z)
$$

Таким образом можно рассматривать течение в прямоугольной трубе, в которой на одной из стенок, при $y=0$, ставится условия прилипания, на других стенках, при $y=1$, $z=0$, $z=L_z/2$, - условия непротикания и проскальзывания.  

Для решения уравнений Навье-Стокса использован тот же численный метод, что и в круглой трубе.
